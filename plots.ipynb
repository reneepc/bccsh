```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
import pandas as pd
from numpy import transpose
```

```python
in_5 = pd.read_csv('entrada/5-processos', sep=' ',header=None)
in_50 = pd.read_csv('entrada/50-processos', sep=' ',header=None)
in_500 = pd.read_csv('entrada/500-processos', sep=' ',header=None)

fcfs_5 = pd.read_csv('saida/fcfs5/comp1-1', sep=' ',header=None)
fcfs_50 = pd.read_csv('saida/fcfs50/comp1-1', sep=' ',header=None)
fcfs_500 = pd.read_csv('saida/fcfs500/comp1-1', sep=' ',header=None)

srtn_5 = pd.read_csv('saida/srtn5/comp1-1', sep=' ',header=None)
srtn_50 = pd.read_csv('saida/srtn50/comp1-1', sep=' ',header=None)
srtn_500 = pd.read_csv('saida/srtn500/comp1-1', sep=' ',header=None)

in_ = [in_5, in_50, in_500]
fcfs = [fcfs_5, fcfs_50, fcfs_500]
srtn = [srtn_5, srtn_50, srtn_500]
```

```python
def deadline (in_, out):
    a = in_[3] - out[1]
    count = 0
    for i in a:
        if i > 0:
            count += 1
    return count
```

```python
dead_fcfs = []
dead_srtn = []
for i in range(0, 3):
    dead_fcfs.append(deadline(in_[i], fcfs[i]))
    dead_srtn.append(deadline(in_[i], srtn[i]))
```

```python
def plot (total, deadline, kind):
    bars = plt.bar([1, 2], [deadline, deadline]) 
    if (kind == 0):
        bars[0].set_color('skyblue')
        bars[1].set_color('cadetblue')
    elif (kind == 1):
        bars[0].set_color('peachpuff')
        bars[1].set_color('lightsalmon')
        
    plt.ylim(0, total)
    labels = ["FCFS", "SRTN"]
    plt.text(1, deadline + 0.1, deadline)
    plt.text(2, deadline + 0.1, deadline)
    plt.xticks([1, 2], labels)
    plt.show()
```

```python
runs = [5, 50, 500]

for i in range(0, 3):
    plot (runs[i], dead_fcfs[i], 0)
    
    
for i in range(0, 3):
    plot (runs[i], dead_srtn[i], 1)
```
